# Connectivity Arduino Uno R3

```python
import serial
import time

user_ports={
    "matteo": ["/dev/ttyACM0", "/dev/ttyACM1"],
    "mike": ["COM3"],
}
#Per MIKE Verifica la connessione qua che su Linux potresti avere una porta diversa da Windows, ho ipotizzato fosse COM3 come standard
for user, ports in user_ports.items():
    for port in ports: 
        if not port:
            continue
        try:
            arduino=serial.Serial(port=port, baudrate=9600, timeout=1)
            print(f"Connected to {user}'s port")
            break
        except serial.SerialException:
            pass
if arduino is None:
    print("Error. Arduino not found, check in Arduino IDE")
    pass

while(True):
    line=arduino.readline().decode("utf-8").strip()
    if line:
        print(f"Received data: {line}")
    time.sleep(1)
```
